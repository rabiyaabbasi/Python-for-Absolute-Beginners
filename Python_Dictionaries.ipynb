Dictionaries Entries --> Key (Lable) : Value (Information)

1. Syntax with Single Entry Dictionary

Andy's email address

Key: Andy

Value: andy@gmail.com

To create a disctionary we use curly brackets {} and to connect key and value we use colon (:)

```python
#Example: Andy's Email Address

{"Andy": "andy@gmail.com"}
```

2. Multi-Entry Dictionary

In multi-entry dictionary, we place key-value pairs inside curly braces {}, separated by commas.

```python
#Example multi-entry dictionary

{"Andy": "andy@gmail.com",
"Ashley": "ashley@gmail.com",
"Ann": "ann@gmail.com"}
```

3. Accessing Dictionary Values

To access a value in a dictionary, store it in a variable and use key inside the square brackets

```python
#Example accessing dictionary values

emails = {"Andy": "  andy@gmail.com",
"Ashley": "ashley@gmail.com",
"Ann": "ann@gmail.com"}
```

```python
#To access ashley's email address

emails["Ashley"]
```

4. Dictionaries vs Lists

```python
#List stores value in an order by position

contacts = ["andy@gmail.com", "ashley123@gmail.com", "ann@gmail.com"]

contacts[0]
```

```python
#Dictionay uses meaningful labels called keys to store values

emails["Ann"]
```

5. Rules for keys and values

Values can be of any data type and can repeat. 

Keys must be of immutable data type and must be unique.

```python
#Various data types for values in a dictionary

siblings_age = {"Andy": 25, "Ashely": 22, "Ann": 27}
has_car = {"Andy": True, "Ashely": False, "Ann": True}

print(siblings_age)
print(has_car)  
```

```python
#Keys must be unique

{
    "Ashley": "email@gmail.com",
    "Ashley": "email2@gmail.com"
}
```

6. Adding and Removing Items

```python
#Add an entry
emails = {"Andy": "andy@gmail.com",
"Ashley": "ashley@gmail.com",
"Ann": "ann@gmail.com"}

emails["Will"] = "will@gmail.com"

emails
```

```python
#remove an entry from a dictionary
emails = {'Andy': 'andy@gmail.com',
 'Ashley': 'ashley@gmail.com',
 'Ann': 'ann@gmail.com',
 'Will': 'will@gmail.com'}

emails.pop("Ann")
emails

```

7. Dictionary methods

```python
#get all keys using .keys() method with for loop

for name in emails.keys():
    print(name)
    
```

```python
#get all values from a dictionary using .values() method with for loop

for email in emails.values():
    print(email)
```

```python
#keys and values together using .items() method with for loop   

for name, email in emails.items():
    print(name, email)
```

8. Find who owns the email id

```python
#loop through key-value pairs in a dictionary
contacts = {
    "John": "john@gmail.com",
    "Jane": "jane@gmail.com",
    "Andy": "andy@gmail.com"
}

for name, email in contacts.items():
    if email == "jane@gmail.com":
        print("This email belongs to:", name)
```
