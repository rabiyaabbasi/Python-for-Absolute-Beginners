Python Math Operators

Four most common operators:

1. '+' for addition

2. '-' for subtraction

3. '*' for multiplication

4. '/' for division

```python
#Example:
2+3
```

```python
x = 5-4
y = 2*6
z = 5/2

print(x,y,z)
```

Exponentiation (power)

Python uses ** (double star) for exponents.

```python
2**2
```

Floor division

// (double slash), called floor division --> keeps whole number

```python
13 // 3
```

Modulus (remainder)

To get the remainder, use %, called the modulus operator.

```python
6 % 2
```

```python
13 % 2
```
