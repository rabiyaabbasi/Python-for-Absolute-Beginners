Checking the type with type()

```python
x = 10
y = 3.5
name = "Hello"
is_active = False

print(type(x))
print(type(y))
print(type(name))
print(type(is_active))
```

Type Casting

```python
#Example: Converting to int using int()

a = 5.7

b = int(a)

print(b)
print(type(b))
```

```python
#Example: Converting string to an integer using int()

num = "10"
converted_num = int(num)

print(converted_num)
print(type(converted_num))

```

```python
#Example: Converting to float using float()

x =10
y= float(x)

print(y)
print(type(y))
```

```python
#Example: Convert numeric strings to float

price = "19.99"
new_price = float(price)

print(new_price)
print(type(new_price))
```

```python
#Example: Converting to str using str()

age = 25
age_text = str(age)

print(age_text)
print(type(age_text))

print("I am " + age_text + " years old")
```
