Modules and Imports

Built-in Module 

Math

```python
import math

sq_root = math.sqrt(16)

print(sq_root)
```

Different ways to import module

1 Import entire module

```python
import math

pi_value = math.pi

print(pi_value)
```

2 Import a specific funcion

```python
from math import sqrt

print(f"The square root of {25} is {sqrt(25)}")
```

3 Import Multiple Functions

```python
from math import sqrt, pow
print(f"{sqrt(9)}")
print(f"{pow(2, 3)}") # 2 raise to power 3
```

Import with an Alias

```python
import math as m

print(f"{m.sqrt(36)}")
```

Lets create our own module and then we will use that here: calculator.py

```python
import calculator

print(f"{calculator.add(10, 5)}")
print(f"{calculator.subtract(10, 5)}")
```

Importing Specific Functions from the module we created

```python
from calculator import add 

print(f"{add(10,3)}")
```

Important Rule: What Runs When You Import? (6:30 – 7:15)

When you import a module, all the code at the top level runs automatically.

That’s why we often use this special line in module file:

if __name__ == "__main__":


This prevents code from running when the file is imported.
