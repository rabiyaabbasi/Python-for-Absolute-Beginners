Variable Naming Rules

Variable naming rule 1 — No leading numbers

```python
my_3_fav_fruits = ["mango", "apple", "kiwi"]
print(my_3_fav_fruits)
```

Variable naming rule 2 — Only letters, digits, and underscores

```python
my_3_fav_fruits = ["mango", "apple", "kiwi"]
```

Variable naming rule 3 — No spaces

```python
my_3_fav_fruits = ["mango", "apple", "kiwi"]
```

Variable naming rule 4 — Case sensitivity

```python
my_3_fav_fruits = ["mango", "apple", "kiwi"]
My_3_fav_fruits = ["mango", "apple", "kiwi"]
MY_3_FAV_FRUITS = ["orange", "banana", "grape"]
print(my_3_fav_fruits)
print(My_3_fav_fruits)
print(MY_3_FAV_FRUITS)
```
