While Loop Syntax

```python
#Example 1

count = 1

while count <= 5:
    print(count)
    count += 1

print("Done!")
```

While vs For Loops

```python
for count in range(1, 6):
    print(count)

print("Done!")
```

The Classic Use Case: Keep Asking Until Correct

```python
password = ""

while password != "secret123":
    password = input("Enter password: ")

print("Access granted!")
```

Cleaner Pattern: while True + break

```python
while True:
    password = input("Enter password: ")

    if password == "secret123":
        break

print("Access granted!")
```
