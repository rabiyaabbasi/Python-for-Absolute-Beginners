Python Nested Loops

```python
#Example 1: Repeating Code (Not ideal)

numbers = [1,2,3]

letters = ["a", "b", "c"]

#Use single loop

for number in numbers:
    print(f"{number}{letters[0]}", end=" ")
    print(f"{number}{letters[1]}", end=" ")
    print(f"{number}{letters[2]}", end=" ")
    print()

```

```python
#Example: Using nested loop (better approach)
numbers = [1,2,3]

letters = ["a", "b", "c"]

for number in numbers:
    for letter in letters:
        print(f"{number}{letter}", end= " ")
    print()
```
