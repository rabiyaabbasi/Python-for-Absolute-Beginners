Python User Input

Built-in function input() for user input

```python
#Example:
name = input("Enter your name:")
print(f"my name is {name}")
```

Rule 1: Input is always string

```python
age = input("Enter your age: ")
print(f"your age is {age} years old")
```

```python
#Example: math input

age = input("Enter your age: ")
print(f"Next year you will be {age+1}")
```

Converting Input Using Type Casting

```python
age = int(input("Enter your age: "))
print(f"Next year you will be {age+1}")
```

Be careful

```python
#Example: To avoid confusion
age = int(input("Enter your age: "))
print(f"Next year you will be {age}+1") # treated as string
```

Multiple inputs

```python
First_name = input("Enter first name: ")
Last_name = input("Enter last name: ")

Full_name = First_name + ' '+ Last_name

print(f"Full name is {Full_name}")
```

Convert the input if you want to add two numbers

```python
#Sum of two numbers (without conversion, gives wrong answer)
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")

print(f"The result is {num1 + num2}")
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print(f"The result is {num1 + num2}")
```

Inputs with Floats

```python
price = float(input("Enter the price: "))
tax = float(input("Enter tax percentage: "))

total = price + (price * tax / 100)
print(f"Total price is {total}")
```
