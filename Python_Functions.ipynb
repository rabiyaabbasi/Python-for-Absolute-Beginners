Python Functions Syntax

In Python, we create a function using the keyword:  def

Then we write a descriptive function name.

Then we add parentheses ().

Inside these parentheses, we define the inputs the function needs.

Separated by a comma.(no need if one input)

Finally, we end the first line with a colon :

That first line is called the function header.

Enter and there is indented line to write a code

Then we write code and in called function body 

```python
#Example --> Version 1 
unit = 'feet squared'

def calculate_rectangle_area(length, width):
    area = length * width
    print(area, unit)


calculate_rectangle_area(14, 10) #function call
```

Name in function header --> parameters --> calculate_rectangle_area(length, width)

The vales you pass while function calling --> arguments --> calculate_rectangle_area(14, 15)

Calcuating the area --> Function execution

Single Responsibility Principle

--> one function should do one job

```python
#Example --> Version 2
#unit1 = 'feet squared'

def calculate_rectangle_area1(length, width):
    area1 = length * width

area1 = calculate_rectangle_area1(14, 15)
print(area1)
```

Namespaces: Global vs Local

--> track variable names and theri values

Global names space --> created outside function and stored globally

unit = "feet squared" --> global variables, global name space also store function names

Local namespace --> when function is called, temporary namespace is created -->stores:

--> parameters (length, width)

--> variables created inside the function (area)

When function ends, local namespace destroys --> Fix add return

Return Statement

```python
unit = 'feet squared'

def calculate_rectangle_area(length, width):
    area = length * width
    return area

area = calculate_rectangle_area(14, 10) #global variable
print(area, unit)
```

```python
unit = 'feet squared'

def calculate_rectangle_area(length, width):
    area = length * width
    return area

result = calculate_rectangle_area(14, 10) 
print(result, unit)
```
