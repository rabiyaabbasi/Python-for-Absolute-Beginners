Python Conditionals

--> if

--> else

--> elif

--> nested conditionals

--> the ternary operator

--> and finally, the difference between = and ==,

```python
#Example : if statement

age = 18

if age >= 18:
    print("you are allowed to vote")
```

```python
#Example: Lets add Else

age = 16

if age >=18:
    print("You are allowed to vote")
else:
    print("you are not allowed to vote")
    
```

```python
#Example: Use elif for multiple conditions

score = 75

if score >= 90:
    print("Grade:A")
elif score >= 80:
    print("Grade:B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

Nested conditionals

Placing an if statement inside another if statement

```python
age = 20
has_id = True


if age >= 18:
    if has_id:
        print("you can enter")
    else:
        print("you need an ID to enter")
else:
    print("you are too young to enter")
```

The ternary operator (one-line ifâ€“else)

```python
#Example --> Simple way

age = 18

if age >= 18:
    status = "Adult"
else:
    status = "minor"

print(status)
```

```python
#Example: Alternative: using the ternary opeartor

status = "Adult" if age >= 18 else "minor"
print(status)
```

Assignment "=" operator vs Comparison "==" operator

```python
#Example: Assignment Operator "="

x = 5

print(x)
```

```python
#Example: Comparison operator "=="
#Lets check if x is equal to 5 --> x==5

x = 5

if x==5:
    print("x is equal to 5")
```
